## Visualization 1: Workforce Composition

The first step is to understand the overall workforce before analyzing employee attrition.
Gender distribution provides a quick overview of employee composition and helps identify whether the workforce is balanced.

```python
# Employee Gender Distribution
gender_counts = df["Gender"].value_counts()
colors = ["#2F6690", "#8FB8DE"]   # Professional blue shades
plt.figure(figsize=(7,7))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.82,
    wedgeprops=dict(width=0.45, edgecolor="white")
)
plt.title(
    "Employee Gender Distribution",
    fontsize=16,
    fontweight="bold",
    color="#2C3E50",
    pad=20
)
plt.tight_layout()
plt.show()
```

## Visualization 2: Employee Distribution by Department

Understanding the number of employees in each department provides a clear picture of the company's workforce.
This helps identify which departments have the largest workforce before analyzing employee attrition.

```python
# Employee Distribution by Department

dept = df["Department"].value_counts()
colors = ["#2563EB", "#F97316", "#22C55E"]
plt.figure(figsize=(8,4))
bars = plt.barh(dept.index, dept.values, color=colors, height=0.6)
plt.gca().invert_yaxis()
plt.title("Employee Distribution by Department", fontsize=14, weight="bold")
plt.xlabel("Number of Employees")

for bar in bars:
    plt.text(bar.get_width()+5,
             bar.get_y()+0.2,
             int(bar.get_width()))

plt.show()
```

## Visualization 3: Employee Age Distribution by Attrition

### Why I selected this chart

I used a histogram because it shows how employee ages are distributed and allows me to compare employees who stayed with those who left.
### Insight
This visualization helps identify whether employee attrition is more common within certain age groups.
### Business Recommendation
If younger or older employees show higher attrition, the company can design targeted retention strategies for those age groups.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Age",
    hue="Attrition",
    bins=15,
    palette=["#2563EB", "#EF4444"],
    alpha=0.8
)
plt.title("Employee Age Distribution by Attrition", weight="bold")
plt.xlabel("Age")
plt.ylabel("Employees")
plt.show()
```

## Visualization 4: Monthly Income by Attrition
### Why I selected this chart
I used a box plot because it compares the salary distribution of employees who stayed and those who left. It also highlights the median, spread, and potential outliers.

### Insight
This chart helps determine whether employee attrition is associated with differences in monthly income.
### Business Recommendation
If employees with lower salaries are leaving more frequently, the company should review its compensation strategy and employee benefits.

```python
plt.figure(figsize=(7,5))
sns.boxplot(
    data=df,
    x="Attrition",
    y="MonthlyIncome",
   palette=["#06B6D4", "#EC4899"]
)
plt.title("Monthly Income by Attrition", weight="bold")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.show()
```

## Visualization 5: Correlation Between Numerical Features
### Why I selected this chart
I used a correlation heatmap to examine the relationships between numerical variables. It helps identify which features have strong positive or negative correlations.
### Insight
The heatmap provides a quick overview of how employee characteristics such as age, income, total working years, and years at the company are related.
### Business Recommendation
Understanding these relationships can help HR identify important factors that influence employee experience and support better workforce planning.

```python
num = df.select_dtypes(include="number")

plt.figure(figsize=(10,8))

sns.heatmap(
    num.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Heatmap", fontsize=14, weight="bold")

plt.show()
```

## Visualization 6: Years at Company vs Monthly Income

### Why I selected this chart

I used a scatter plot to explore the relationship between employees' years at the company and their monthly income. This chart helps identify trends, patterns, and possible differences between employees who stayed and those who left.

### Insight

The scatter plot shows that monthly income generally increases as employees spend more years at the company. It also highlights the distribution of employees based on their attrition status.

### Business Recommendation

The company should continue providing career growth opportunities and competitive salary progression to encourage long-term employee retention and improve employee satisfaction.

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=df,
    x="YearsAtCompany",
    y="MonthlyIncome",
    hue="Attrition",
    palette=["#6366F1", "#F97316"]
)
plt.title("Years at Company vs Monthly Income", weight="bold")
plt.xlabel("Years at Company")
plt.ylabel("Monthly Income")
plt.show()
```

## Visualization 7: Overall Employee Attrition

### Why I selected this chart
I used a pie chart to show the overall proportion of employees who stayed and those who left the company. This provides a simple summary of the employee attrition rate.
### Insight
The chart shows that most employees stayed with the company, while a smaller percentage left. This gives an overall picture of employee retention before concluding the analysis.
### Business Recommendation
Although the overall attrition rate is relatively low, the company should continue monitoring employee turnover and focus on retaining high-performing employees.

```python
attrition = df["Attrition"].value_counts()
colors = ["#FF7D00", "#E6AACE"]
plt.figure(figsize=(6,6))
plt.pie(
    attrition,
    labels=attrition.index,
    autopct="%1.1f%%",
    colors=colors,
    startangle=90,
    explode=[0, 0.08],
    wedgeprops={"edgecolor":"white"}
)
plt.title("Overall Employee Attrition", weight="bold")
plt.show()
```
